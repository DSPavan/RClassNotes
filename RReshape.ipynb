```r
molted = melt(airquality,id.vars=c("Month","Day"),measured.vars=c("Ozone","Solar.R","Wind"),variable="measured")
str(molted)
molted[sample(nrow(molted),10),] # Random sampling
# VAriable Renamed
```

```r
# casting
```

```r
library("dplyr")
dcast(molted,Month+Day~measured)

# In reshape2, cast() has been replaced by two functions: acast() and dcast().
```

```r
dcast(molted, Month~measured, fun.aggregate=mean, na.rm=TRUE)
```

```r
dcast(molted,measured~.,fun.aggregate=mean,, na.rm=TRUE)
```
