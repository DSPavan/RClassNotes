```r
# sorting examples using the mtcars dataset
attach(mtcars)

# sort by mpg
newdata1 <- mtcars[order(mpg),] 

# sort by mpg and cyl
newdata2 <- mtcars[order(mpg, cyl),]

#sort by mpg (ascending) and cyl (descending)
newdata3 <- mtcars[order(mpg, -cyl),] 

detach(mtcars)
```

```r
attach(mtcars)
aggdata <-aggregate(mtcars, by=list(cyl,vs), 
  FUN=mean, na.rm=TRUE)
print(aggdata)
detach(mtcars)
```

```r
attach(mtcars)
aggdata <-aggregate(mtcars, by=list(cyl.group=cyl,gear.group=gear), FUN=mean, na.rm=TRUE)
print(aggdata)
```
