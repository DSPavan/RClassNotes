```r
library(e1071)
data(iris)
#head(iris, 3)
iris.part = subset(iris ) # , Species != 'setosa'
iris.part$Species = factor(iris.part$Species)
head(iris.part)
```

```r

iris.part = iris.part[, c(1,2,5)]
# iris.part = iris.part[, c(3,4,5)]
fit = svm(Species ~ ., data=iris.part, 
          type='C-classification', kernel='linear')
plot(fit, iris.part)
```
