```r
# Principal Component Analysis (PCA) is a useful technique for exploratory data analysis, 
# allowing you to better visualize the variation present in a dataset with many variables.
head(mtcars)
```

```r
mtcars.pca <- prcomp(mtcars[,c(1:7,10,11)], center = TRUE,scale. = TRUE)

summary(mtcars.pca)
```

```r
# LOOK INTO Proportion of Variance : PC1 0.62, PC2 0.23 = 0.84 means 84% of data is important between Pc1 to Pc9

# MPG, CYL are important ( 1, 2)

str(mtcars.pca)
```

```r
plot(mtcars.pca)
```

```r
biplot(mtcars.pca, scale = 0)
```

```r
fviz_pca_ind(mtcars.pca,
             col.ind = "cos2", # Color by the quality of representation
             gradient.cols = c("#00AFBB", "#E7B800", "#FC4E07"),
             repel = TRUE     # Avoid text overlapping
             )
```

```r
fviz_pca_var(mtcars.pca,
             col.var = "contrib", # Color by contributions to the PC
             gradient.cols = c("#00AFBB", "#E7B800", "#FC4E07"),
             repel = TRUE     # Avoid text overlapping
             )
```

```r
fviz_pca_biplot(mtcars.pca, repel = TRUE,
                col.var = "#2E9FDF", # Variables color
                col.ind = "#696969"  # Individuals color
                )
```
