```r
DF<-data.frame(Maths=c(80, 90, 95), Science=c(85, 85, 80), English=c(60, 70, 40), Music=c(55, 45, 50))
DF
```

```r
df2 <- prcomp(DF[,1:4], scale = FALSE)
summary(df2)
```

```r
biplot(df2, scale = 0)
```

```r
library("factoextra")
var <- get_pca_var(df2)
var
```

```r
head(var$contrib)
```

```r
fviz_cos2(df2, choice = "var", axes = 1:2)
```

```r
library("corrplot")
corrplot(var$cos2, is.corr=FALSE)
```
